```python
try:
    for i in range(data.shape[0]):
        #存储关键词的位置
        with open(keywords_filepath, 'a', newline='') as f:
            # 不理解，为什么要一直重新创建啊啊啊
            builder = appbuilder.AppBuilderClient(app_id)
            conversation_id = builder.create_conversation()
            #以下部分为处理选题信息数据
            #将事实层、价值原型等信息合并到text
            if (value[i] != 0 and fact[i] != 0):
                text = value[i] +"。"+ fact[i]
            elif (value[i] == 0 and fact[i] != 0):
                text = fact[i]+"。"
            elif (value[i] != 0 and fact[i] == 0):
                text = value[i]+"。"
            else:
                print(labels[i], "输入有误！！")
                text = "无"
            if keywords[i] != 0:
                text = keywords[i] +"。"+ text
            if ad[i] != 0:
                text = text + "。"+ad[i]
            if len(text)>1200:
                text = text[0:1200]

            #运行
            try:
                msg = builder.run(conversation_id, query=text)
            except Exception:
                out = "0"
                print("异常！异常！")
                writer = csv.writer(f)
                writer.writerow([id[i], labels[i], out])
                f.close()
            else:
                out = msg.content.answer
                writer = csv.writer(f)
                writer.writerow([id[i],labels[i], out])
                print('第', (i + 1), '条信息写入成功。书名：', labels[i],'  |  关键词：',out)
                f.close()
except Exception:
    f.close()

print('信息写入完毕，共',(i+1),'条数据。')
```
